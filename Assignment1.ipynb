```python
import pandas as pd
import datetime
df = pd.read_csv('Europe Sales Records.csv')
#Task 1 Dataset overview
#total number of rows and columns
df.shape
```

```python
#Name od all columns
df.columns
```

```python
#datatypes of each columns
df.dtypes
```

```python
#missing values
df.isnull()
```

```python
#to check duplicates
df.duplicated()
```

```python
#dropping duplicates
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
#total revenue generated by each country
total_revenue= df.groupby('Country')['Total Revenue'].sum()
print(total_revenue)
```

```python
#top 10 highest revenue country
print(total_revenue.sort_values(ascending=False).head(10))
```

```python
#total profit genertaed by each country
total_profit = df.groupby('Country')['Total Profit'].sum()
print(total_profit)
```

```python
#top 5 countries with highest profit
print(total_profit.sort_values(ascending=False).head(5))
```

```python
#top 5 countries having lowest profit
print(total_profit.sort_values(ascending=False).tail(5))
```

```python
#toal units sold per item type
unit_sold_itype= df.groupby('Item Type')['Units Sold'].sum()
print(unit_sold_itype)
```

```python
# item type which is sold the maximum
print(unit_sold_itype.sort_values(ascending=False).head(1))
```

```python
#total profit per item type
total_profit_itype= df.groupby('Item Type')['Total Profit'].sum()
print(total_profit_itype.sort_values(ascending=False).head(1))
```

```python
#total revenue per sales channe;
total_revenue = df.groupby('Sales Channel')['Total Revenue'].sum()
print(total_revenue)
```

```python
#total profit per sales channel
total_profit_channel = df.groupby('Sales Channel')['Total Profit'].sum()
print(total_profit_channel)
```

```python
# best sales channel
best_channel = total_profit_channel.head(1)
print(best_channel)
```

```python
#order count
count_of_order = df['Order Priority'].value_counts()
print(count_of_order)
```

```python
# average profit per order priority
avg_profit = df.groupby('Order Priority')['Total Profit'].mean()
print(avg_profit)
```

```python
print(avg_profit.head(1))
```

```python
#correlation between total revenue and total profit
correlation = df['Total Revenue'].corr(df['Total Profit'])
print(correlation)
```

```python
#check  whether highest profit generates high revenue
check = df.groupby('Total Revenue')['Total Profit'].mean()
print(check)
print(check.sort_values(ascending=False).head(1))
```

```python
#conversion of order id date to datetime format using datetime module
convert_order_date= pd.to_datetime(df['Order Date'])
print(convert_order_date)
```

```python
#total revenue of each month
total_revenue_month = df.groupby(convert_order_date.dt.month)['Total Revenue'].sum()
print(total_revenue_month)
```

```python
#number of the month that generated highest revenue
print(total_revenue_month.sort_values(ascending=False).head(1))
```

```python
#number of the month generated lowest revenue
print(total_revenue_month.sort_values(ascending=False).tail(1))
```

```python
#business summary report having total revenue,total profit,best perofrming country best selling item type total order processed and best sales channels
print("Total Revenue:", df['Total Revenue'].sum())
print("Total Profit:", df['Total Profit'].sum())
print("Best Performing Country:", total_revenue.sort_values(ascending=False).head(1).index[0])
print("Best Selling Item Type:",unit_sold_itype.sort_values(ascending=False).head(1))
print("Total Order Processed:", df['Order ID'].nunique())
print("Best Sales Channel:",best_channel)
```
